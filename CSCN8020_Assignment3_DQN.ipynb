## Task Summary

The task is to train a PyTorch Deep Q-Network to control the fixed-base Unitree G1 left elbow in the approved Gymnasium environment. The DQN receives four observation values: current elbow angle, current elbow velocity, goal angle, and angle error. It outputs three Q-values for the discrete actions: decrease target, hold target, and increase target.

Training goals are sampled from -0.8 rad to +0.8 rad. Final greedy evaluation uses epsilon = 0.0 on four benchmark goals: -0.8, -0.4, +0.4, and +0.8 rad, with five episodes per goal.

## Student-Written DQN Components

- `src/dqn/q_network.py`: 4-input, 64-unit ReLU, 64-unit ReLU, 3-output Q-network.
- `src/dqn/replay_buffer.py`: bounded random replay buffer.
- `src/dqn/agent.py`: epsilon-greedy action selection, online and target networks, Bellman targets, Huber loss, gradient clipping, target-network synchronization, checkpoint save/load.
- `src/train_dqn.py`: headless training loop and metrics logging.
- `src/evaluate_dqn.py`: greedy DQN and rule-based baseline evaluation.
- `src/render_dqn_policy.py`: saved-checkpoint MuJoCo viewer demonstration.

## Reproducibility Commands

Run from the repository root in WSL Ubuntu:

```bash
source .venv/bin/activate
export PYTHONPATH=src
python -m compileall src tests
python -m unittest tests.test_dqn_components -v
python src/smoke_test_dqn.py
```

Train the two epsilon-decay configurations:

```bash
python src/train_dqn.py --config-name config_a --epsilon-decay 0.995 --episodes 500 --output-dir results/config_a --model-path models/config_a_dqn.pt
python src/train_dqn.py --config-name config_b --epsilon-decay 0.985 --episodes 500 --output-dir results/config_b --model-path models/config_b_dqn.pt
```

Evaluate the selected checkpoint greedily:

```bash
python src/evaluate_dqn.py --policy dqn --checkpoint models/selected_dqn.pt --episodes-per-goal 5 --output-csv results/selected_dqn_evaluation.csv
```

Render the selected policy:

```bash
python src/render_dqn_policy.py --checkpoint models/selected_dqn.pt --goals -0.8 0.8
```
